Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading Dataset for data transparency

```python
file_path = '/content/Road_Accident_Data_2015_2022.csv'
data = pd.read_csv(file_path)
```

```python
# Plotting a bar chart to compare Total Accidents, Total Fatalities, and Total Injuries by Organization
plt.figure(figsize=(10, 6))

# Plot bars for each metric
data.plot(kind='bar', x='Organization', y=['Total Accidents', 'Total Fatalities', 'Total Injuries'],
          figsize=(10, 6), color=['blue', 'red', 'green'])
# Adding title and labels
plt.title("Comparison of Total Accidents, Fatalities, and Injuries by Organization")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")

plt.legend(title="Metrics")
plt.tight_layout()
plt.show()
```

Read the data from main dataset

```python
df = pd.read_csv("/content/BD_Accident_Data - Sheet1.csv")
```

Plotting the total road accidents

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Total Road Accidents"], marker='o')
plt.title("Total Road Accidents Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Road Accidents")
plt.xticks(df["Year"])
plt.grid()
plt.show()
```

Comparing the total accident, total fatalities, total accidents

```python
# Set the bar width
bar_width = 0.25

# Set the position of bar on X axis
r1 = range(len(df["Year"]))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(r1, df["Total Road Accidents"], color='b', width=bar_width, edgecolor='grey', label='Total Road Accidents')
plt.bar(r2, df["Total Fatalities"], color='r', width=bar_width, edgecolor='grey', label='Total Fatalities')
plt.bar(r3, df["Total Injuries"], color='g', width=bar_width, edgecolor='grey', label='Total Injuries')

# Add labels and title
plt.xlabel('Year', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(df["Year"]))], df["Year"])
plt.ylabel('Count', fontweight='bold')
plt.title('Comparison of Total Road Accidents, Fatalities, and Injuries')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

Calculate the fatality rate using total fatalities and total road accident

```python
# Calculate the fatality rate
df['Fatality Rate (%)'] = (df['Total Fatalities'] / df['Total Road Accidents']) * 100

# Display the results
print(df[['Year', 'Total Road Accidents', 'Total Fatalities', 'Fatality Rate (%)']])
```

```python
# Calculate fatality rate
df['Fatality_Rate'] = (df['Total Fatalities'] / df['Total Road Accidents']) * 100

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Fatality_Rate"], marker='o')
plt.title("Fatalities Rate over Year")
plt.xlabel("Year")
plt.ylabel("Fatality Rate (%)")
plt.xticks(df["Year"])
plt.grid()
plt.show()
```

Caculating the Fatalities rate during festival

```python
df['Festival Fatalities'] = df['Fatalities during Festival']

# Calculate the ratio of festival fatalities to total fatalities
df['Festival Fatality Ratio'] = (df['Festival Fatalities'] / df['Total Fatalities']) * 100

# Display the results
print(df[['Year', 'Total Fatalities', 'Festival Fatalities', 'Festival Fatality Ratio']])
```

```python
# Calculate the ratio of festival fatalities to total fatalities
df['Festival Fatality Ratio'] = (df['Festival Fatalities'] / df['Total Fatalities']) * 100

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Festival Fatality Ratio"], marker='o')
plt.title("Fatalities Rate over Year")
plt.xlabel("Year")
plt.ylabel("Fatality Rate (%)")
plt.xticks(df["Year"])
plt.grid()
plt.show()
```

```python
# Data for 2022 and 2023
years = [2022, 2023]
buses = [941, 1011]
trucks = [1653, 1548]
cng_vehicles = [419, 370]
cars = [469, 337]
motorcycles = [1929, 1640]
rickshaws = [770, 905]
three_wheelers = [568, 450]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, buses, label='Buses', color='b', alpha=0.7)
plt.bar(years, trucks, bottom=buses, label='Trucks', color='g', alpha=0.7)
plt.bar(years, cng_vehicles, bottom=[b + t for b, t in zip(buses, trucks)], label='CNG Vehicles', color='r', alpha=0.7)
plt.bar(years, cars, bottom=[b + t + c for b, t, c in zip(buses, trucks, cng_vehicles)], label='Cars', color='y', alpha=0.7)
plt.bar(years, motorcycles, bottom=[b + t + c + ca for b, t, c, ca in zip(buses, trucks, cng_vehicles, cars)], label='Motorcycles', color='m', alpha=0.7)
plt.bar(years, rickshaws, bottom=[b + t + c + ca + m for b, t, c, ca, m in zip(buses, trucks, cng_vehicles, cars, motorcycles)], label='Rickshaws', color='c', alpha=0.7)
plt.bar(years, three_wheelers, bottom=[b + t + c + ca + m + r for b, t, c, ca, m, r in zip(buses, trucks, cng_vehicles, cars, motorcycles, rickshaws)], label='Three-Wheelers', color='purple', alpha=0.7)

plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Vehicle Types Involved in Road Accidents (2022 vs. 2023)')
plt.xticks(years)
plt.legend()
```
